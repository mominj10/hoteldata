```python
df.drop(['company','agent'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#to select adr<5000 as its a big outlier
df=df[df['adr']<5000]
```

```python
#.value_counts(): occurrences of unique values in the specified column, in this case, the 'is_canceled' column.
#normalize=True:represent proportions (percentages) rather than raw counts. Setting normalize=True means that the result will be the relative frequencies of each unique valu
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
#plotiing the cancelled and not cancelled reservations
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.4)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1= sns.countplot(x='hotel', hue= 'is_canceled',data=df,palette='Blues')
legend_labels=ax1.get_legend_handles_labels()
plt.title('Reservation in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in City and Resort Hotel', fontsize=25)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled',data=df)
plt.legend(['Not cancelled','Canceled'])
plt.ylabel('Number of reservations')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=20)

# group 'month' and 'adr' are columns in your DataFrame
sns.barplot(x='month',y= 'adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())


plt.show()
```

```python
canceldata=df[df['is_canceled']==1]
top10=canceldata['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top10,autopct='%.2f',labels=top10.index)
plt.show()
```

```python
canceldata_adr=canceldata.groupby('reservation_status_date')[['adr']].mean()
canceldata_adr.reset_index(inplace=True)
canceldata_adr.sort_values('reservation_status_date',inplace=True)

notcanceldata_adr=notcanceldata.groupby('reservation_status_date')[['adr']].mean()
notcanceldata_adr.reset_index(inplace=True)
notcanceldata_adr.sort_values('reservation_status_date',inplace=True)
```

```python
canceldata_adr=canceldata_adr[(canceldata_adr['reservation_status_date']>'2016')& (canceldata_adr['reservation_status_date']<'2017')]
notcanceldata_adr=notcanceldata_adr[(notcanceldata_adr['reservation_status_date']>'2016')& (notcanceldata_adr['reservation_status_date']<'2017')]
plt.figure(figsize=(20,8))
plt.title('Average daily rate',fontsize=20)
plt.plot(notcanceldata_adr['reservation_status_date'],notcanceldata_adr['adr'],label='Not cacnceled')
plt.plot(canceldata_adr['reservation_status_date'],canceldata_adr['adr'],label='Not cacnceled')
```
